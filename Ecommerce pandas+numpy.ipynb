```python
# 10- What Hour recives high demand for ordering?
df["hours"]=df["InvoiceTime"].apply(lambda x:x[3:5])
df.groupby("hours")["Quantit"].sum().sort_values(ascending=False)
```

```python
df["hours"]=df["hours"].astype("Int32")
```

```python
# 11- What's the Time of the day that contributes to high Order Values?
def part_of_day(value):
    if value<12:
        return "Morning"
    elif value>=12 and value<18:
        return "Afternoon"
    else:
        return "Evening"
df["OrderTimeOfDay"]=df["hours"].apply(part_of_day)
df.groupby("OrderTimeOfDay")["Quantit"].sum().sort_values(ascending=False)
```
